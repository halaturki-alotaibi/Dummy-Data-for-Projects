```python
from faker import Faker
import pandas as pd
import random
from datetime import timedelta

# Create a Faker object
fake = Faker()

# Number of rows
num_rows = 1500

# القيم الممكنة لبعض الأعمدة
device_types = ['Desktop', 'Mobile', 'Tablet']
browsers = ['Chrome', 'Safari', 'Firefox', 'Edge']
referrers = ['Google', 'Instagram', 'Twitter', 'TikTok','Snapchat']
utm_sources = ['ads', 'influencer', 'seo', None]
pages = ['/home', '/about', '/products', '/contact', '/blog', '/login']
saudi_cities = ["Riyadh","Jeddah","Mecca","Medina","Dammam","Khobar","Abha","Tabuk","Hail","Buraidah"]

# List to collect data
data = []

for _ in range(num_rows):
    is_registered = random.choice([True, False])
    signup_date = fake.date_time_between(start_date='-1y', end_date='now') if is_registered else None
    visit_datetime = fake.date_time_between(start_date='-30d', end_date='now')
    page_views = random.randint(1, 10)
    pages_visited = random.sample(pages, k=page_views if page_views <= len(pages) else len(pages))
    session_duration = random.randint(30, 1200)  # in seconds

    data.append({
        'user_id': fake.uuid4(),
        'is_registered': is_registered,
        'visit_datetime': visit_datetime,
        'session_id': fake.uuid4(),
        'device_type': random.choice(device_types),
        'browser': random.choice(browsers),
        'referrer': random.choice(referrers),
        'page_views': page_views,
        'session_duration_sec': session_duration,
        'country': 'Saudi Arabia',  # The country is stable
        'city': random.choice(saudi_cities),  # City from the list
        'signup_date': signup_date,
        'user_type': random.choice(['New', 'Returning']),
        'conversion': random.choice([True, False]),
        'pages_visited': pages_visited,
        'utm_source': random.choice(utm_sources)
    })

# Convert it to a DataFrame
df = pd.DataFrame(data)

# Show first 5 rows
df.head()
```

```python
df.to_csv("saudi_website_visitors_data.csv", index=False)
```
